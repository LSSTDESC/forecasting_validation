```python
import emcee
from getdist import plots, MCSamples, loadMCSamples
import matplotlib.pyplot as plt
```

```python
reader = emcee.backends.HDFBackend("modified_res_600_y1.h5")
samples = reader.get_chain(flat=True)
samples_getdist_res600 = MCSamples(samples=samples, names=['om',
                                                              'ob', 
                                                              'h',
                                                              'sigma8',
                                                              'n_s',
                                                              'w0',
                                                              'wa'], 
                                                      labels=[r'\Omega_\mathrm{m}',
                                                              r'\Omega_\mathrm{b}',
                                                              r'h',
                                                              r'\sigma_8',
                                                              r'w_0',
                                                              r'w_\mathrm{a}'],
                                                              label='low res',
                                                              )

samples_getdist_res600.removeBurn(remove=.3)
```

```python
reader = emcee.backends.HDFBackend("benchmark_y1.h5")
samples = reader.get_chain(flat=True)
print(samples.shape)
samples_getdist_benchmark = MCSamples(samples=samples, names=['om',
                                                              'ob', 
                                                              'h',
                                                              'sigma8',
                                                              'n_s',
                                                              'w0',
                                                              'wa'], 
                                                      labels=[r'\Omega_\mathrm{m}',
                                                              r'\Omega_\mathrm{b}',
                                                              r'h',
                                                              r'\sigma_8',
                                                              r'w_0',
                                                              r'w_\mathrm{a}'],
                                                              label='benchmark')
samples_getdist_benchmark.removeBurn(remove=.3)

```

```python
import matplotlib.pyplot as plt

g = plots.getSubplotPlotter(analysis_settings={'fine_bins':20, 'fine_bins_2D':20})
g.triangle_plot([samples_getdist_benchmark, samples_getdist_res600], filled=True)

plt.savefig("contour.pdf")
```
